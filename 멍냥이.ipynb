```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# intelModule은 자주쓰는 함수를 미리 만들어 놓은 클래스입니다.
df1 = pd.read_csv("dog Intelligence.csv", encoding='ISO-8859-1')
df1.info()
df1.head(3)
```

```python
# intelModule은 자주쓰는 함수를 미리 만들어 놓은 클래스입니다.
df2 = pd.read_csv("dog_breeds.csv")
df2.info()
df2.head(3)
```

```python
# df1과 df2를 'Breed' 열을 기준으로 병합하여 공통된 값만 남기기
common_breeds = pd.merge(df1, df2, on='Breed', how='inner')

print("Common Breeds:")
print(common_breeds)
```

```python
# weight 열의 평균값을 계산하여 새로운 열 추가
common_breeds['Weight Average (lbs)'] = (common_breeds['weight_low_lbs'] + common_breeds['weight_high_lbs']) / 2

print("Common Breeds with Weight Average:")
print(common_breeds)
```

```python
# 크기별로 나누기 위한 조건 함수
def classify_size(weight_average):
    if weight_average < 10:
        return 'Small'
    elif 10 <= weight_average < 25:
        return 'Medium'
    else:
        return 'Large'

# 'Weight Average (lbs)' 열을 기준으로 크기별로 나누기
dfS = common_breeds[common_breeds['Weight Average (lbs)'] < 10]
dfM= common_breeds[(common_breeds['Weight Average (lbs)'] >= 10) & (common_breeds['Weight Average (lbs)'] < 25)]
dfL = common_breeds[common_breeds['Weight Average (lbs)'] >= 25]

print("Small Breeds:")
print(dfS)

print(dfM)
print(dfM)

print("\nLarge Breeds:")
print(dfL)
```


소형견의 common health problems

```python
from collections import Counter

# 열의 값을 가져와서 쉼표로 구분된 리스트로 변환
values = dfS['Common Health Problems'].str.split(', ')

# 모든 단어를 하나의 리스트로 모음
all_words = [word for sublist in values.dropna() for word in sublist]

# 단어별 빈도수 계산
word_counts = Counter(all_words)

# 가장 빈도수가 높은 단어 10개 추출
top_words = word_counts.most_common(15)

# 결과 출력
for word, count in top_words:
    print(f"{word}: {count}")


plt.figure(figsize=(10, 6))
plt.bar(*zip(*top_words))
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Words')
plt.xticks(rotation=45)
plt.show()
```

```python
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt

# 모든 단어별 빈도수 합산
all_word_counts = Counter()
for word, count in top_words:
    all_word_counts[word] += count

# WordCloud 생성 및 저장
wc = WordCloud(font_path='malgun', width=800, height=800, scale=2.0, max_font_size=150, background_color='white')
gen = wc.generate_from_frequencies(all_word_counts)

plt.figure(figsize=(10, 8))
plt.imshow(gen, interpolation="bilinear")
plt.axis("off")
plt.title("Common Health Problems")
plt.show()
```
